```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = "https://weather.com/en-IN/weather/tenday/l/a5f0fe2ff9a40acc9ce62d67cd99439a71cde78cc0c5c1fbf6da052bef4cdba9"
response = requests.get(url,headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36"})
print(response)
soup = BeautifulSoup(response.text,'html.parser')
```

```python
days = soup.find_all('h2',class_="DetailsSummary--daypartName--kbngc")

for i in days:
  print(i.text)
```

```python
H_temp = soup.find_all('span', class_='DetailsSummary--highTempValue--3PjlX')

high_temp = []

for i in H_temp:
  high_temp.append(i.text)

print(high_temp)
```

```python
L_temp = soup.find_all('span',class_='DetailsSummary--lowTempValue--2tesQ')

low_temp=[]

for i in L_temp:
  low_temp.append(i.text)

print(low_temp)
```

```python
weather_con = soup.find_all('div', class_='DetailsSummary--condition--2JmHb')

condition = []

for i in weather_con:
  condition.append(i.text)

print(condition)
```

```python
rain_per = soup.find_all('div',class_='DetailsSummary--precip--1a98O')

rain_percent = []

for i in rain_per:
  rain_percent.append(i.text)

print(rain_percent)
```

```python
wind = soup.find_all('span',class_='Wind--windWrapper--3Ly7c undefined')
wind_ = []
for i in wind:
  wind_.append(i.text)

print(wind_)
```

```python
weather_data = {"day name":days, "Day high temprature":high_temp, "Day low temprature":low_temp, "weather_conditions":condition, "rain_percentage":rain_percent, "wind_speed":wind_}
df = pd.DataFrame(weather_data)
df
```

```python
df[['wind_direction','wind_speed']]=df['wind_speed'].str.split(' ',expand=True)
df
```

```python
df.to_csv("weather.csv")
```
